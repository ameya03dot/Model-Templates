```python
transactions = []  #Right format for Apriori Algorithm

for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])    
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions = transactions,min_support = 0.003,min_confidence =0.2,min_length = 2,min_lift = 3,max_length =2)
```

```python
results = list(rules)
results
```
